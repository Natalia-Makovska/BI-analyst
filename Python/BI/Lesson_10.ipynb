```python
fig, ax = plt.subplots()

xs = np.arange(-100, 101, 10)
ys = xs ** 3 

ax.plot(
xs,
ys,
linewidth=2,
color='red',
linestyle='--',
alpha=1,
marker='o',
label='line 1'
);

ax.plot(
xs,
ys * -1,
linewidth=2,
color='blue',
linestyle='--',
alpha=1,
marker='o',
label='line 2'
);

ax.legend()
```

Load dataset of US births between 2000 - 2014. Show the total number of births per year in a line plot.

```python
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2018/2018-10-02/us_births_2000-2014.csv')
```

```python
df
```

```python
births_per_year = df.groupby('year').agg(total_births=('births', 'sum')) 
                    # назва колонки  з якої колонки, як агрегувати
births_per_year
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(births_per_year)
ax.set_xlabel('Year')
ax.set_ylabel('Births')
ax.set_title('US births by year')
#ax.set_ylim((births_per_year.total_births.min(), births_per_year.total_births.max()))
```

```python
plt.plot(births_per_year.index, births_per_year.values)
```

```python
plt.plot(births_per_year.values, births_per_year.index)
```
