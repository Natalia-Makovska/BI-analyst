```python
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-08/ipf_lifts.csv')
df
```

2.Find the record of each gender and each division in each of the 3 lifts (bench press,squat and deadlift).

```python
df_record = df.groupby(by=['division', 'sex'])[['best3squat_kg','best3bench_kg', 'best3deadlift_kg']].agg(['max'])
df_record
```

3.Count the number of victories of each contestant, considering a victory as winning 1stplace. Store the results in a secondary DataFrame.

```python
df_victory = df. groupby('place').get_group('1')
df_group = df_victory.groupby(by=['division','name', 'sex']).agg(number_of_victories =('place', 'count'))
df_group
```

4.Use the dataframe from ex. 3 to show, for each combination of division and gender, the contestant with most victories among contestants who participated in that division at some time.

```python
df_group = df_victory.groupby(by=['division','sex', 'date', 'name']).agg(number_of_victories =('place', 'count'))
df_group_sort = df_group.sort_values('number_of_victories', ascending=False)
df_group_sort
```
