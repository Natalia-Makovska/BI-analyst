# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 8], [5, 1, 3, 2])
ax.plot([5, 1, 3, 2], [1, 2, 3, 8])
fig
```

```python
ax.plot([1, 2, 3, 4], [5, 1, 3, 2])
fig
```

```python
plt.plot([1, 2, 3, 8], [5, 1, 3, 2])
```

```python
fig, ax = plt.subplots()

def f(x):
    return x ** 2


xs = range(-100, 101)

ys = [f(x) for x in xs]


ax.plot(xs, ys);
```

```python
fig, ax = plt.subplots()

xs = np.array(range(-100, 101))
ys = xs ** 2

ax.plot(xs, ys);
```
